```python
publisher_counts = data['publisher'].value_counts()

# Choose the top N publishers to display
top_publishers = 50  # You can adjust this number based on your preferences

# Identify the top N publishers and group the rest into 'Other'
top_publisher_data = publisher_counts.head(top_publishers)
other_publisher_count = publisher_counts.iloc[top_publishers:].sum()
top_publisher_data['Other'] = other_publisher_count

# Plot the bar chart
plt.figure(figsize=(12, 8))
top_publisher_data.plot(kind='bar', color='skyblue', width=0.5)
plt.title(f'Top {top_publishers} Publishers and Others')
plt.xlabel('Publisher')
plt.ylabel('Number of Books')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
top_publishers = data['publisher'].value_counts().nlargest(5).index.tolist()
```

```python
for publisher in top_publishers:
    filtered_df = data[data['publisher'] == publisher]
    category_distribution = filtered_df['category'].value_counts(normalize=True) * 100
    threshold = 5  # Set a threshold percentage for combining into 'Other'
    small_proportions = category_distribution[category_distribution < threshold]
    category_distribution['Other'] = small_proportions.sum()
    category_distribution = category_distribution.drop(small_proportions.index)

# Plot the pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(category_distribution, labels=category_distribution.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Set3.colors)
    plt.title(f'Category Distribution for {publisher} Books')
    plt.show()
```

```python
top_publishers = data['publisher'].value_counts().nlargest(6).index.tolist()  # Choose top 2 for example
```

```python
df_top_publishers = data[data['publisher'].isin(top_publishers)]

# Create a grouped bar plot for the relationship between 'category' and the top N publishers
category_publisher_counts = df_top_publishers.groupby(['category', 'publisher']).size().unstack(fill_value=0)
category_publisher_counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Relationship between Category and Top Publishers')
plt.xlabel('Category')
plt.ylabel('Count')
plt.legend(title='Publisher')
plt.show()
```

```python
corr = data[['height', 'weight', 'price', 'width']]
co_ma = corr.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(co_ma, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```
